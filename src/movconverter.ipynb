```python

def convert(input_directory):
    for filename in os.listdir(input_directory):
        if filename.endswith(".mov"):
            input_file = os.path.join(input_directory, filename)
            output_file = os.path.join(input_directory, os.path.splitext(filename)[0] + ".mp4")

            # Convert the input video file to MP4 format
            (
                ffmpeg
                .input(input_file)
                .output(output_file)
                .run(overwrite_output=True)
            )

            # Remove the original MOV file
            os.remove(input_file)
```

```python
convert("extended_subset/hvor/")
convert("extended_subset/mange/")
convert("extended_subset/nej/")
convert("extended_subset/nu/")
convert("extended_subset/tid/")
```

```python
def rename(path, word):
    i = 0
    for file in glob.glob(path):
        # Split the path into a list using the / delimiter
        path_list = path.split("/")

        # Extract the word between the second to last and the last /
        path = path_list[:-1]
        newfile =  os.path.join(path, word+str(i)+".mp4")
        os.rename(file, newfile)
        i += 1
        print(newfile)
```

```python
import os

def rename_files_in_folder(folder_path):
    """
    Renames all files in a folder as the parent folder's name plus an iterator.
    """
    # Get the parent folder's name
    parent_folder_name = os.path.basename(os.path.dirname(folder_path))

    # Loop over all files in the folder
    for i, filename in enumerate(os.listdir(folder_path)):
        # Construct the new file name with the parent folder's name and an iterator
        file_path = os.path.join(folder_path, filename)
        new_filename = f"{parent_folder_name}{i+1}{os.path.splitext(filename)[1]}"
        new_file_path = os.path.join(folder_path, new_filename)

        # Rename the file to the new file name
        os.rename(file_path, new_file_path)
        #print(new_file_path)
```

```python
rename_files_in_folder("extended_subset/fordi/")
rename_files_in_folder("extended_subset/hvor/")
rename_files_in_folder("extended_subset/mange/")
rename_files_in_folder("extended_subset/nej/")
rename_files_in_folder("extended_subset/nu/")
rename_files_in_folder("extended_subset/tid/")
```
